2.4. Модифицировать функцию 2.2. так, чтобы рештку можно было замыкать по строкам и по столбцам (с помощью соответсвующих параметров функции). Проверить результат работы функции визуально.

```python
import networkx as nx
from matplotlib import pyplot as plt

def visualize_grid_graph(n, m, closed_rows=False, closed_columns=False):
    G = nx.grid_2d_graph(n, m)

    if closed_rows:
        for row in range(n):
            G.add_edge((row, 0), (row, m - 1))
    if closed_columns:
        for col in range(m):
            G.add_edge((0, col), (n - 1, col))

    plt.figure(figsize=(6,6))

    nx.draw(G,
        node_color='lightblue',
        with_labels=True,
        node_size=600)

n = 5
m = 4
visualize_grid_graph(n, m, closed_rows=True, closed_columns=False)
plt.show()
```

Для построенного граафа 1.2 цветка получить:

3.1. Релизовать функцию построения 1,2 цветка. Функция принимает один целочисленный параметр n - количество шагов построения.

```python
import networkx as nx
import matplotlib.pyplot as plt

def fractal_graph(n, G=None):
    if n == 1:
        G = nx.complete_graph(3)
    else:
        if G is None:
            G = fractal_graph(n-1)
        edges_to_add = []
        G_edges = list(G.edges())
        for edge in G_edges:
            node1, node2 = edge
            new_node = len(G.nodes())
            G.add_node(new_node)
            edges_to_add.append((node1, new_node))
            edges_to_add.append((new_node, node2))
        G.add_edges_from(edges_to_add)
    return G

def visualize_fractal_graph(n):
    G = fractal_graph(n)
    plt.figure(figsize=(8, 8))
    # Создание словаря степеней узлов
    node_degrees = dict(G.degree())

    # Вычисление размера узла в зависимости от его степени
    node_size = [v * 100 for v in node_degrees.values()]

    # Построение графа с размерами узлов, зависящими от степени
    pos = nx.spring_layout(G)
    nx.draw(G, pos, node_color='lightblue', node_size=node_size, font_size=8)

    # Добавляем метки с информацией о степени каждого узла
    labels = {node: f"{node}\nDegree: {G.degree(node)}" for node in G.nodes()}
    nx.draw_networkx_labels(G, pos, labels=labels, font_size=8, font_color='black', verticalalignment='center', horizontalalignment='right')
    # Вычисляем информацию о графе
    num_nodes = G.number_of_nodes()
    num_edges = G.number_of_edges()
    avg_degree = sum(dict(G.degree()).values()) / num_nodes

    # Выводим информацию о графе в заголовок
    plt.title(f"Fractal Graph (n={n}), Nodes: {num_nodes}, Edges: {num_edges}, Avg Degree: {avg_degree:.4f}")
    plt.show()

n = 4
G = fractal_graph(n)
visualize_fractal_graph(n)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def visualize_fractal_graph(n):
    G = fractal_graph(n)
    plt.figure(figsize=(8, 8))
    # Вычисление коэффициента кластеризации для каждого узла
    clustering_coeffs = nx.clustering(G)
    # Определение цвета узла в зависимости от его коэффициента кластеризации
    node_colors = [clustering_coeffs[node] for node in G.nodes()]
    # Вычисление размера узла в зависимости от его коэффициента кластеризации
    node_size = [clustering_coeffs[node] * 1000 for node in G.nodes()]

    # Построение графа с размерами узлов, зависящими от коэффициента кластеризации
    pos = nx.spring_layout(G)
    nx.draw(G, pos, node_color=node_colors,cmap='viridis', node_size=node_size, font_size=8)

    # Добавляем метки с информацией о степени и коэффициенте кластеризации каждого узла
    labels = {node: f"{node}\nClustering Coefficient: {nx.clustering(G, node):.4f}" for node in G.nodes()}
    nx.draw_networkx_labels(G, pos, labels=labels, font_size=8, font_color='black', verticalalignment='center', horizontalalignment='right')
    # Добавляем цветовую шкалу
    sm = plt.cm.ScalarMappable(cmap='viridis')
    sm.set_array([])
    plt.colorbar(sm, label='Clustering Coefficient')
    avg_clustering = nx.average_clustering(G)
    plt.title(f"Fractal Graph (n={n}), Average Clustering Coefficient: {avg_clustering:.4f}")
    plt.show()

n = 4
G = fractal_graph(n)
visualize_fractal_graph(n)
```

4\.1 Распределение степеней узлов в виде списка картежей вида: (степень, количество\_узлов\_с\_данной\_степенью), картежи в списке отсортированы по возрастанию степени.

4\.3 График распределения степеней узлов, в котором по вертикальной оси откладыватся выборочная частота. Отобразить график в трех масштабах осей: (x, y), (x, ln y), (ln x, ln y).

```python
import numpy as np
def degree_distribution(G):
    # Получаем распределение степеней узлов
    degree_hist = nx.degree_histogram(G)

    # Преобразуем список в список кортежей (степень, количество узлов с данной степенью)
    distribution = [(degree, count) for degree, count in enumerate(degree_hist) if count != 0]

    # Сортируем список кортежей по возрастанию степени
    distribution.sort(key=lambda x: x[0])

    return distribution
n = 4
G = fractal_graph(n)
# Пример использования
distribution = degree_distribution(G)
print("Degree Distribution:")
for degree, count in distribution:
    print(f"Degree {degree}: {count} nodes")
degrees, counts = zip(*distribution)

# Вычисляем выборочную частоту
total_nodes = sum(counts)
frequency = [count / total_nodes for count in counts]

# Построение графика в трех масштабах осей
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Масштаб (x, y)
axs[0].plot(degrees, frequency, 'o-')
axs[0].set_title('Degree Distribution (x, y)')
axs[0].set_xlabel('Degree')
axs[0].set_ylabel('Sample Frequency')
axs[0].grid(True)

# Масштаб (x, ln y)
axs[1].plot(degrees, np.log(frequency), 'o-')
axs[1].set_title('Degree Distribution (x, ln y)')
axs[1].set_xlabel('Degree')
axs[1].set_ylabel('ln(Sample Frequency)')
axs[1].grid(True)

# Масштаб (ln x, ln y)
axs[2].plot(np.log(degrees), np.log(frequency), 'o-')
axs[2].set_title('Degree Distribution (ln x, ln y)')
axs[2].set_xlabel('ln(Degree)')
axs[2].set_ylabel('ln(Sample Frequency)')
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

Для построенного граафа 1.2 цветка получить:

5\.1\. Два произвольных узла и кратчайший путь между ними.

5\.2\. Визуализировать кратчайший путь, полученный в предыдущем пункте, используя связи другого цвета и другой толщины.

5\.3\. Средний кратчайший путь для всего графа.

5\.4\. Диаметр графа.

```python
import random
def visualize_fractal_graph(n):
    G = fractal_graph(n)
    plt.figure(figsize=(8, 8))

    pos = nx.spring_layout(G)
    nx.draw(G, pos, node_color='lightblue', font_size=8)

    # Добавляем метки с информацией о степени каждого узла
    labels = {node: f"{node}" for node in G.nodes()}
    nx.draw_networkx_labels(G, pos, labels=labels, font_size=8, font_color='black', verticalalignment='center', horizontalalignment='right')
    # Выбираем случайные узлы
    node1 = random.choice(list(G.nodes()))
    node2 = random.choice(list(G.nodes()))

    # Находим кратчайший путь между узлами
    shortest_path = nx.shortest_path(G, node1, node2)

    # Нарисовать кратчайший путь
    path_edges = [(shortest_path[i], shortest_path[i+1]) for i in range(len(shortest_path)-1)]
    nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=5)
    # Вычисляем средний кратчайший путь для всего графа
    avg_shortest_path = nx.average_shortest_path_length(G)
    # Вычисляем диаметр графа
    diameter = nx.diameter(G)
    # Выводим информацию о графе и о кратчайшем пути в заголовок
    plt.title(f"Fractal Graph (n={n}), Shortest Path: {shortest_path}, Average Shortest Path: {avg_shortest_path:.4f}, Diameter: {diameter}")
    plt.show()

n = 4
G = fractal_graph(n)
visualize_fractal_graph(n)
```

## 7.

7\. Релизовать функцию построения u,v цветка. Функция принимает на вход 3 параметра: u – длина 1го пути при замене связи; v – длина 2го пути при замене связи; n - количество шагов построения.

```python
import networkx as nx
import matplotlib.pyplot as plt

def u_v_flower_graph(u, v, n, G=None):
    if n == 1:
        G = nx.cycle_graph(4)
    else:
        if G is None:
            G = u_v_flower_graph(u, v, n-1)
        edges_to_add = []
        G_edges = list(G.edges())
        for edge in G_edges:
            node1, node2 = edge
            new_nodes = len(G.nodes()), len(G.nodes()) + 1

            # Добавляем новые узлы
            G.add_nodes_from(new_nodes)

            # Добавляем пути длины u и v вместо существующей связи
            G.add_edge(node1, new_nodes[0])
            G.add_edge(new_nodes[0], node2)
            G.add_edge(node1, new_nodes[1])
            G.add_edge(new_nodes[1], node2)

            # Удаляем существующую связь
            G.remove_edge(node1, node2)
    return G

def visualize_u_v_flower_graph(u, v, n):
    G = u_v_flower_graph(u, v, n)
    plt.figure(figsize=(8, 8))

    # Создание словаря степеней узлов
    node_degrees = dict(G.degree())

    # Вычисление размера узла в зависимости от его степени
    node_size = [v * 100 for v in node_degrees.values()]

    # Построение графа с размерами узлов, зависящими от степени
    pos = nx.spring_layout(G)
    nx.draw(G, pos, node_color='lightblue', node_size=node_size, font_size=8)

    # Добавляем метки с информацией о степени каждого узла
    labels = {node: f"{node}\nDegree: {G.degree(node)}" for node in G.nodes()}
    nx.draw_networkx_labels(G, pos, labels=labels, font_size=8, font_color='black', verticalalignment='center', horizontalalignment='right')

    # Вычисляем информацию о графе
    num_nodes = G.number_of_nodes()
    num_edges = G.number_of_edges()
    avg_degree = sum(dict(G.degree()).values()) / num_nodes

    # Выводим информацию о графе в заголовок
    plt.title(f"u,v Flower Graph (u={u}, v={v}, n={n}), Nodes: {num_nodes}, Edges: {num_edges}, Avg Degree: {avg_degree:.4f}")
    plt.show()

u = 1
v = 2
n = 2
visualize_u_v_flower_graph(u, v, n)
```
